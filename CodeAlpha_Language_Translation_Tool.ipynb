```python
import gradio as gr
from deep_translator import GoogleTranslator
```

```python
def translate_text(text, source_lang, target_lang):
    language_mapping = {
        "Auto-Detect": "auto",
        "Urdu": "ur",
        "Arabic": "ar",
        "Hindi": "hi",
        "English": "en",
        "French": "fr",
        "Spanish": "es",
        "German": "de",
    }

    source_code = language_mapping[source_lang]
    target_code = language_mapping[target_lang]

    translator = GoogleTranslator(source = source_code, target = target_code)
    translated_text = translator.translate(text)

    return translated_text
```

```python
ui_inputs = [
    gr.Textbox(label = "Enter text to translate", lines = 3),
    gr.Dropdown(choices = ["Auto-Detect", "English", "Spanish", "Arabic", "French", "German", "Urdu", "Hindi"], label = "Souce Language", value = "Auto-Detect"),
    gr.Dropdown(choices = ["Auto-Detect", "English", "Spanish", "Arabic", "French", "German", "Urdu", "Hindi"], label = "Target Language", value = "Hindi")
]
```

```python
ui_outputs = [
    # gr.Textbox(label = "Translated Text", lines = 3)
    gr.Textbox(
    label="Translated Text",
    show_copy_button=True
)
]
```

```python
app = gr.Interface(
    fn = translate_text,
    inputs = ui_inputs,
    outputs = ui_outputs,
    title = "AI Global Translation App",
    description = "Enter text and select your languages to see the translation instantly."
)
```

```python
app.launch()
```
